```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/cardio_train_new_new.csv')
df.head()
```

```python
print(df.columns)
```

```python
x=df['smoke']
y=df['cardio']
print(x)
print(y)
```

```python
df.isnull().sum()
```

```python
correlation_matrix=df.corr()
print(correlation_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.xlabel('smoke')
plt.ylabel('cardio')
plt.title('smoke vs cardio')
plt.show()
```

```python
plt.bar(x,y)
plt.xlabel('smoke')
plt.ylabel('cardio')
plt.title('smoke vs cardio')
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('smoke')
plt.ylabel('cardio')
plt.title('smoke vs cardio')
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
log_reg=LogisticRegression()
log_reg.fit(x_train.values.reshape(-1,1),y_train)
```

```python
y_pred=log_reg.predict(x_test.values.reshape(-1,1))
print(y_pred)
```

```python
accuracy_score_log_reg=accuracy_score(y_pred,y_test)
print(accuracy_score_log_reg)
```

```python
svm=SVC()
svm.fit(x_train.values.reshape(-1,1),y_train)
```

```python
y_pred=svm.predict(x_test.values.reshape(-1,1))
print(y_pred)
```

```python
accuracy_score_svm=accuracy_score(y_pred,y_test)
print(accuracy_score_svm)
```

```python
decision_tree=DecisionTreeClassifier()
decision_tree.fit(x_train.values.reshape(-1,1),y_train)
```

```python
y_pred=decision_tree.predict(x_test.values.reshape(-1,1))
print(y_pred)
```

```python
accuracy_score_tree=accuracy_score(y_pred,y_test)
print(accuracy_score_tree)
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train.values.reshape(-1,1),y_train)
```

```python
y_pred=knn.predict(x_test.values.reshape(-1,1))
print(y_pred)
```

```python
accuracy_score_knn=accuracy_score(y_pred,y_test)
print(accuracy_score_knn)
```

```python
random_forest=RandomForestClassifier()
random_forest.fit(x_train.values.reshape(-1,1),y_train)
```

```python
y_pred=random_forest.predict(x_test.values.reshape(-1,1))
print(y_pred)
```

```python
accuracy_score_random=accuracy_score(y_pred,y_test)
print(accuracy_score_random)
```
